The US Government's Alternative Fuels Data Center collects records of electric vehicle (EV) charging infrastructure, including charging ports and station locations, as well as sales of electric vehicles. With the EV market rapidly evolving, understanding trends in charging facilities and sales is essential to inform strategic planning.

As a data scientist working for a leading EV charging network operator, you recognize the potential in this data and start wrangling and visualizing the aggregated yearly data. 

This yearly data captured in December of each year encompasses a record of EV charging port installations and station localities spanning roughly ten years, capturing both public and private charging environments. 
___

### The Data
&nbsp;

`private_ev_charging.csv`

| Variable   | Description                                          |
|------------|------------------------------------------------------|
| `year` |  Year of data collection |
| `private_ports`| The number of available charging ports owned by private companies in a given year  |
| `private_station_locations`   | The number of privately owned station locations for EV charging

___

`public_ev_charging.csv`
 
| Variable   | Description                                          |
|------------|------------------------------------------------------|
| `year` |  Year of data collection  |
| `public_ports`| The number of available charging ports under public ownership in a given year  |
| `public_station_locations`   | The number of publicly owned station locations for EV charging

___

The sales information is available for each model and year in the `ev_sales.csv` file:

| Variable   | Description                                          |
|------------|------------------------------------------------------|
| `Vehicle` |  Electric vehicle model |
| `year`| Year of data collection |
| `sales`   | The number of vehicles sold in the US

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df_private = pd.read_csv('private_ev_charging.csv')
df_public = pd.read_csv('public_ev_charging.csv')
df_sales = pd.read_csv('ev_sales.csv')
```

```python
# Explore ev sales
# Display the first few rows
print('Display the first few rows of ev_sales')
print(df_sales.head())

# Display basic statistics
print('Display basic statistics of ev_sales')
print(df_sales.describe())

# Check the data types and non-null counts
print('Check the data types and non-null counts of ev_sales')
print(df_sales.info())

# Check shape of the data
print('Check shape of ev_sales')
print(df_sales.shape)

# Explore numeric variables
# Histogram of sales
df_sales['sales'].hist(bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of sales')
plt.show()

# Boxplot
sns.boxplot(y=df_sales['sales'])
plt.title('Boxplot of sales with original scale')
plt.show()

# Boxplot
sns.boxplot(y=df_sales['sales'])
plt.ylim(0, 35000)
plt.title('Boxplot of sales with limited scale')
plt.show()
```

- The distribution of sales is heavily right-skewed, with the vast majority of observations concentrated between 0 and 20,000.
- A very small number of extremely large values (e.g., above 100,000) are present, confirming the presence of extreme outliers.
- These outliers create a long tail and visually compress the bulk of the data, making it harder to interpret the central distribution.
## Outliers and Spread (Boxplot  Original Scale):
- The original-scale boxplot confirms the presence of significant outliers, with at least two extreme values exceeding 140,000.
- The interquartile range (IQR) and the median (~a few thousand) are barely visible in this plot because the y-axis is stretched due to the outliers.
- This makes it hard to evaluate the real distribution of the majority of the data points.
## Focused Insight (Boxplot  Limited Scale):
- With the y-axis capped at 35 000, the revised boxplot offers much better visibility into the main distribution.
- The majority of data points lie below 10,000, with a dense cluster of moderate outliers ranging from 10,000 to ~30,000.
- The central 50% of data (IQR) is between approximately 2,000 and 8,000, and the median sits around 4,0005,000.

```python
# Vehicles sold in 2018
df_sales_filtered = df_sales[df_sales['year'] == 2018]
total_sales = df_sales_filtered['sales'].sum()
ev_sales_2018 = total_sales
print(f'Total number of ev sold in 2018 is {ev_sales_2018}')

# Distribution of sales by years
df_total_sales_by_year = df_sales.groupby('year')['sales'].sum().reset_index()
df_total_sales_by_year.rename(columns={'sales': 'total_sales'}, inplace=True)
print('Distribution of sold ev by years.')
print(df_total_sales_by_year)

# Plot sales by years
sns.barplot(data=df_total_sales_by_year, x='year', y='total_sales', color='skyblue')
plt.title('EV Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

## Strong Upward Growth
- The data clearly shows a consistent year-over-year increase in electric vehicle (EV) sales from 2011 to 2018. This suggests growing adoption, greater availability, and possibly stronger policy or market incentives supporting EVs during that period.
## 2018 as a Peak Year
- The year 2018 marked the highest sales on record, reaching approximately 361,315 units sold. This represents a massive leap from the previous years, highlighting a major acceleration in EV uptake.
## Slight Dip in 2019
- While 2019 remained a strong year overall, sales slightly declined compared to 2018  from ~361k to ~326k. This could indicate:
- Market saturation in key regions,
- Waiting for newer models/releases,
- Policy shifts or subsidy phaseouts,
- Or global events influencing automotive markets.
## 10x Growth in 7 Years
- From just 17,763 units in 2011 to over 360,000 in 2018, the EV market saw a 20x increase in less than a decade  an extraordinary growth trajectory.

```python
# Trends for private ports, public ports and sales
# Aggregate total EV sales by year
df_ev_sales_by_year = df_sales.groupby('year')['sales'].sum().reset_index()

# Make sure all datasets are structured similarly
df_private_trimmed = df_private[['year', 'private_ports']]
df_public_trimmed = df_public[['year', 'public_ports']]

# Rename columns to 'value' and add a source column to each
df_ev_sales_by_year['type'] = 'sales'
df_ev_sales_by_year.rename(columns={'sales': 'value'}, inplace=True)

df_private_trimmed['type'] = 'private_ports'
df_private_trimmed.rename(columns={'private_ports': 'value'}, inplace=True)

df_public_trimmed['type'] = 'public_ports'
df_public_trimmed.rename(columns={'public_ports': 'value'}, inplace=True)

# Combine all three data sets
df_combined = pd.concat([df_ev_sales_by_year, df_private_trimmed, df_public_trimmed])

# Plot using fig, ax
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(data=df_combined, x='year', y='value', hue='type', marker='o', ax=ax)

# Optional polish
ax.set_title('Trends of EV Sales, Private & Public Charging Ports')
ax.set_ylabel('Count')
ax.set_xlabel('Year')
plt.tight_layout()
plt.show()
trend = 'same'
```

## Charging Infrastructure Growth (Public vs Private):
- Public Charging Ports have shown a steady and consistent increase across the entire timeline, particularly accelerating after 2018. By 2022, theyve more than quintupled compared to early years, indicating strong public investment or policy backing.
- Private Charging Ports are also growing but at a slower and more gradual pace, with relatively flat growth from 2014 to 2019, followed by a slight uptick after 2020. This may reflect more selective or region-specific private sector expansion.
## EV Sales Spike and Fluctuation:
- EV sales rose sharply between 2015 and 2018, peaking at around 360,000 units in 2018, suggesting a major market shift, likely driven by: Increased consumer interest, More available models, Government incentives.
-   Sales dropped slightly in 2019, hinting at a possible market correction, policy change, or supply chain constraint.
-   Sales data ends earlier than infrastructure (likely 2019), indicating that infrastructure continues expanding even without current sales data  perhaps to prepare for future demand or fill earlier gaps.
## Sales vs Infrastructure Alignment:
- Theres a positive correlation between EV sales and the growth of charging infrastructure  as sales increased, so did the number of both private and public ports.
- However, sales volume is much higher than port availability, highlighting the importance of scaling charging networks to avoid bottlenecks as EV adoption grows.

```python
# Distribution of sales by vehicle model
top_models = df_sales.groupby('Vehicle')['sales'].sum().sort_values(ascending=False).head(10).reset_index()
sns.barplot(data=top_models, y='Vehicle', x='sales')
print('List of top models sold')
print(top_models)

plt.title('EV Sales by Vehicle Model')
plt.xlabel('Vehicle')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```
